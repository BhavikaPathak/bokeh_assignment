# ANSWER 2
In Bokeh, glyphs are geometric shapes that represent the data points to be plotted. Glyphs are used to visually display data in various forms, such as circles, squares, lines, bars, etc. Each glyph is associated with a data source, and its properties can be customized to represent different aspects of the data, like size, color, transparency, etc.

To add glyphs to a Bokeh plot, by using the relevant glyph function provided by Bokeh. For example,to create a scatter plot, use the scatter() glyph function.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import random
x = list(range(1, 11))
y = [random.randint(0, 100) for _ in range(10)]

output_notebook() 
plot = figure(title="Simple Scatter Plot", x_axis_label="X-axis", y_axis_label="Y-axis")


plot.scatter(x, y, size=10, color="blue", alpha=0.8)

show(plot)
```

# ANSWER 3
Customize the appearance of a Bokeh plot, including the axes, title, and legend, by accessing various attributes and methods of the plot object. Bokeh provides a wide range of options to tailor the plot to your specific needs.

```python
# EXAMPLE OF TITLE
x = list(range(1, 11))
y = [random.randint(0, 100) for _ in range(10)]

output_notebook() 
plot = figure(title="SCATTER PLOT")
plot.scatter(x, y, size=10, color="blue", alpha=0.8)

show(plot)
```

```python
# EXAMPLE OF AXIS
x = list(range(1, 11))
y = [random.randint(0, 100) for _ in range(10)]

output_notebook() 
plot = figure(x_axis_label="X-axis", y_axis_label="Y-axis")

plot.scatter(x, y, size=10, color="blue", alpha=0.8)

show(plot)
```

```python
# EXAMPLE OF LEGEND
x = list(range(1, 11))
y = [random.randint(0, 100) for _ in range(10)]

output_notebook() 
plot.scatter(x, y, size=10, color="blue", alpha=0.8, legend_label="SCATTER DATA POINTS")
plot.legend.location = "top_left"
plot.legend.title = "Legend Title"
show(plot)
```

A Bokeh server is a Python process that allows to create and deploy interactive web applications and dashboards that update and respond to user input or changes in data in real time. With the Bokeh server, you can build dynamic and interactive plots that go beyond static visualizations, enabling users to interact with the plots and explore data in real time.

To use the Bokeh server to create interactive plots that can be updated in real time,need to follow these steps:

1. Define the Plot and Data: First, create the Bokeh plot(s) and prepare the data you want to visualize.Use Bokeh's glyph functions (e.g., scatter(), line(), etc.) to add data points to the plot.
2. Add Interactivity: To make the plot interactive, add widgets like sliders, buttons, dropdowns, etc., using Bokeh's widget functions (e.g., Slider(), Button(), etc.). These widgets will act as user input elements that trigger updates in the plot.
3. Define Callbacks: TO create Python callback functions that respond to changes in the widgets and update the data or properties of the plot accordingly. These callbacks will be executed when users interact with the widgets.
4. Set Up the Bokeh Server Application: Use the curdoc() function from Bokeh to create the Bokeh document (Bokeh's internal data model for plotting). Then, add the plot and widgets, along with their callbacks, to this document.
5. Run the Bokeh Server: Save your Python script containing the Bokeh server application, and run it from the command line using bokeh serve followed by the filename. This starts the Bokeh server and hosts your interactive web application at a given URL.
